```python
import numpy as np
```

```python
dataset = [10,20,20,30,100]
data = np.array(dataset)
print(np.mean(data))
```

```python
dataset = data - 36
print(dataset)
```

```python
dataset = np.square(dataset)
print(dataset)
```

```python
variance = np.mean(dataset)
print(variance)
```

```python
std_var = np.sqrt(variance)
print(std_var)
```
